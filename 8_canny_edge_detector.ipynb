```python
#import packages
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
#func to display images 
def plt_showimage(title, image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.title(title)
    plt.grid(False)
    plt.show()
```

**Canny Edge Detector**

```python
#read image
image = cv2.imread("/content/drive/MyDrive/OpenCV102 - Basic Image Processing Operations/coins.jpg")

#convert into gray scale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#Slightly blurr this image
blurred = cv2.GaussianBlur(gray,(5,5), 0)

plt_showimage("Original", image)
plt_showimage("Blurred", blurred)
```

```python
#Use canny edge detector
#wide
#mid-range
#tight

wide = cv2.Canny(blurred, 10, 200) #2nd argument is Tupper and 3rd argument is Tlower
mid = cv2.Canny(blurred, 30,150)
tight = cv2.Canny(blurred, 248, 270)


plt_showimage("Wide Edge Map", wide)
plt_showimage("Mid Edge Map", mid)
plt_showimage("Tight Edge Map", tight)
```
