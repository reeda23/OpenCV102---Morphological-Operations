```python
#import packages
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
#function to display image

def plt_showimage(title, image):
    if len(image.shape) == 3:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    plt.imshow(image)
    plt.title(title)
    plt.grid(False)
    plt.show()
```

```python
#read image from the disk

image = cv2.imread("/content/drive/MyDrive/OpenCV102 - Basic Image Processing Operations/coins.jpg")
plt_showimage("Original image", image)
```

```python
#converting image to gray scale and blurring it a little bit to smooth the details inside the coins

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray, (7,7), 0) #zero means cv2 automatically controls the value of sigma
```

```python
plt_showimage("Blurred Image",blurred)
```

**Basic Thresholding**

```python
(T, threshInv) = cv2.threshold(blurred, 190, 255, cv2.THRESH_BINARY_INV)
plt_showimage("Basic Thresholding", threshInv)
```

```python
#using normal thresholding

(T, thresh) = cv2.threshold(blurred, 190, 255, cv2.THRESH_BINARY)
plt_showimage("Normal Thresholding", thresh)
```

```python
#masking 

masked = cv2.bitwise_and(image, image, mask=threshInv)
plt_showimage("Mask", masked)
```

**Otsu Thresholding**

```python
#Otsu thresholding automatically decides the best threshold value

(T, threshInv) = cv2.threshold(blurred, 0, 255, cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)
plt_showimage("OTSU Thresholding", threshInv)
print("Optimal value of T is {}".format(T))
```
