```python
#import packages
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
#func to show image
def plt_showimage(title, image):
    if len(image.shape) == 3:
        image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.title(title)
    plt.grid(False)
    plt.show()
```

```python
#read image
image = cv2.imread("/content/drive/MyDrive/OpenCV102 - Basic Image Processing Operations/adrian.png")
plt_showimage("RGB", image)

#displaying each channel individually
for (name, ch) in zip(("B","G","R"), cv2.split(image)):
    plt_showimage(name,ch)
```

```python
#HSV color space
#read image
image = cv2.imread("/content/drive/MyDrive/OpenCV102 - Basic Image Processing Operations/adrian.png")
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
plt_showimage("HSV", hsv)

#displaying each channel individually
for (name, ch) in zip(("H","S","V"), cv2.split(hsv)):
    plt_showimage(name,ch)
```

```python
#L*a*b* color space

#read image
image = cv2.imread("/content/drive/MyDrive/OpenCV102 - Basic Image Processing Operations/adrian.png")
lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
plt_showimage("L*a*b*", lab)

#displaying each channel individually
for (name, ch) in zip(("L*","a*","b*"), cv2.split(lab)):
    plt_showimage(name,ch)
```

```python
#grayscale 
image = cv2.imread("/content/drive/MyDrive/OpenCV102 - Basic Image Processing Operations/adrian.png")

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt_showimage("Original", image)
plt_showimage("Grayscale", gray)
```
