```python
#import packages
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
#func to display images

def plt_showimage(title, image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.title(title)
    plt.grid(False)
    plt.show()
```

```python
#read image 
image = cv2.imread("/content/drive/MyDrive/OpenCV102 - Basic Image Processing Operations/steve_jobs_card.jpg")
plt_showimage("Original Image",image)
```

```python
#pre-processing of image
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray, (7,7), 0)
```

```python
#applying global thresholding
(T, threshInv) = cv2.threshold(blurred, 230, 255, cv2.THRESH_BINARY_INV)
plt_showimage("Global Thresholding", threshInv)
```

```python
#Otsu thresholding
(T, threshInv) = cv2.threshold(blurred, 0, 255, cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)
plt_showimage("OTSU Thresholding", threshInv)
print(T)
```

**Adaptive Thresholding**

```python
#adaptive thresholding
thresh = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 21, 10) #21 is local subregion and 10 is constant
plt_showimage("Adaptive Thresholding (Mean)", thresh)
```

```python
#adaptive thresholding
thresh = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 21, 5) #21 is local subregion and 10 is constant
plt_showimage("Adaptive Thresholding (Gaussian)", thresh)
```
